```python
import pandas as pd
import os
```

```python
directory = '../data/2016_chunks/'
all_dfs = []  
for filename in os.listdir(directory):
    print(filename)
    df = pd.read_csv(directory+filename, low_memory=True)
    all_dfs.append(df)
```

```python
final_df = pd.concat(all_dfs)
final_df.to_csv('2016_final.csv', index=False)
```
