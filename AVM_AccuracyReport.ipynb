```python
import pandas as pd
import math
import numpy as np
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df = pd.read_csv('../data/train_with_feautures.csv')
df['ratio'] = np.exp(df['logerror'])
df
```

```python
df[['year', 'month', 'logerror','ratio', 'White_Percentage','Black_Percentage','h_l_Percentage','not_h_l_Percentage','Asian_Percentage']]
```

```python
ppe5_lower = math.log(95/100)
ppe5_uppr = math.log(105/100)

ppe10_lower = math.log(90/100)
ppe10_uppr = math.log(110/100)

ppe20_lower = math.log(80/100)
ppe20_uppr = math.log(120/100)
```

```python
df['ppe5']  = (df.logerror >= ppe5_lower ) & ( df.logerror <= ppe5_uppr)
df['ppe10'] = (df.logerror >= ppe10_lower ) & ( df.logerror <= ppe10_uppr)
df['ppe20'] = (df.logerror >= ppe20_lower ) & ( df.logerror <= ppe20_uppr)
df['over'] = df.logerror > 0.0
df['under'] = df.logerror < 0.0
```

```python
df
```

# Overall Results

```python
df[['ppe5', 'ppe10', 'ppe20', 'over', 'under', 'logerror', 'ratio']].mean()
```

```python
df.groupby('year')[['ppe5', 'ppe10', 'ppe20', 'over', 'under', 'logerror', 'ratio']].mean().T
```

# By Census Tract

```python
keep_ids = [x for x,y in df.census_tract.value_counts().items() if y >= 50]
df[df['census_tract'].isin(keep_ids)].groupby('census_tract')[['ppe5', 'ppe10', 'ppe20', 'over', 'under', 'logerror', 'ratio']].mean()['under'].max()
```

# Performance Across Ethnicities

```python
df_performance = df_white.join(df_black.join(df_h_l))
df_performance['Percent_Diff_Black_White'] = (df_performance['> 50% Black'] - df_performance['> 50% White'])*100
df_performance['Percent_Diff_H/L_White']   = (df_performance['> 50% H/L'] - df_performance['> 50% White'])*100

df_performance['Percent_Diff_H/L_White'] = round(df_performance['Percent_Diff_H/L_White'], 4).astype(str) + '%'
df_performance['Percent_Diff_Black_White'] = round(df_performance['Percent_Diff_Black_White'], 4).astype(str) + '%'
df_performance
```
